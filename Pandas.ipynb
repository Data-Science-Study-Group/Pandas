```python
# to convert the date filed to date datatype while importing it 
df_new1 = pandas.read_csv("C:\\Users\\LOKESH\\Desktop\\Datascience\\Analyticspth_python\\py-master\\py-master\\pandas\\5_handling_missing_data_fillna_dropna_interpolate\\weather_data.csv", parse_dates=["day"])
df_new1
```

```python
type(df_new.day[0])
```

```python
# now use day column as index to your table (replace the real one )

df_new1.set_index("day",inplace=True)
```

```python
df_new1
```

```python
new = df_new1.fillna(0)
new
```

```python
# but it's not advisable to fill everything with 0 so we can use dictionary in fillna function

new1 = df_new1.fillna({
                        "temperature":0,
                        "windspeed":0,
                         "event": "No Event"
                      })

new1
```

```python
# but it's not good to let them 0 while calculating mean or something so we use "forwardfill" method
# in "forwardfill" method it will fill the gaps with previous values 

new2 = df_new1.fillna(method="ffill")
new2
```

```python
# we can use "backwordfill" method

new3 = df_new1.fillna(method= "bfill")
new3
```

```python
# we can use "axis" method to fill horizontally too

new4 = df_new1.fillna(method='bfill',axis="columns")  # we can use forwardfill too here but "bfill" is more reasonable here 
new4
```

```python
# we have "limit" here to limit the fill to once or depending on our requirement

new5 = df_new1.fillna(method="ffill", limit=1)
new5

# see the output, here 2017-01-09 won't take the previous value because we limit this to 1 "ffill" 
```

```python
# now we will see interpolate

new6 = df_new1.interpolate()            # this interpolate do that it takes linear values default between the missing values 
new6                                    # but in interpolate there are many methods like liner, polynomial....etc
```

```python
# use "time" method to get results based on datetime

new7 = df_new1.interpolate(method="time")           
new7
```

```python
# dropna method 
# this method will drop all NA values from the dataset

new8 = df_new1.dropna()
new8
```

```python
# but we have to drop all NA value row only rather than at least one NA value 
# in that case we will use "how" method

new9 = df_new1.dropna(how="all")
new9

# 2017-01-09 row has deleted because it has all NAs
```

```python
# how to insert missing dates 

dt = pandas.date_range("01-01-2017","01-11-2017")
idx = pandas.DatetimeIndex(dt)

df_new1 = df_new1.reindex(idx)
df_new1
```

```python
# for exact results now we can use interpolate method

new10 = df_new1.interpolate()
new10
```
